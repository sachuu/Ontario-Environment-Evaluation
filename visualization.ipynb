## Question 1

Compare the mean Regular Unleaded Gasoline price among different cities in 2019. Plot the corresponding bar chart.

```python
df['Date'] = pd.to_datetime(df['Date'])
filteredData = df[(df['Fuel Type'] == "Regular Unleaded Gasoline") & (df['Date'] > '12-31-2018') & (df['Date'] <= '12-31-2019')]

ottawaMean = filteredData['Ottawa'].mean()
torontoWMean = filteredData['Toronto West/Ouest'].mean()
torontoEMean = filteredData['Toronto East/Est'].mean()
londonMean = filteredData['London'].mean()
sudburyMean = filteredData['Sudbury'].mean()
stMarieMean = filteredData['Sault Saint Marie'].mean()
thunderBayMean = filteredData['Thunder Bay'].mean()
northBayMean = filteredData['North Bay'].mean()
timminsMean = filteredData['Timmins'].mean()

df1 = pd.DataFrame({'Cities':['Ottawa', 'TorontoW', 'TorontoE', 'London', 'Sudbury', 'StMarie', 'Thunderbay', 'Northbay', 'Timmins'], 'Weekly Retail Price of Gasoline':[ottawaMean, torontoWMean, torontoEMean, londonMean, sudburyMean,stMarieMean, thunderBayMean, northBayMean, timminsMean]})
ax = df1.plot.bar(x='Cities', y='Weekly Retail Price of Gasoline', rot=45)
```
